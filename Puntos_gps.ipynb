```python
# Instala las bibliotecas necesarias
!pip install geopy folium

# Importa las bibliotecas
from geopy.geocoders import Nominatim
import folium

# Ingresa la direcci√≥n que deseas buscar
direccion = "1600 Amphitheatre Parkway, Mountain View, CA"

# Obt√©n las coordenadas usando Geopy
geolocator = Nominatim(user_agent="my_geocoder")
ubicacion = geolocator.geocode(direccion)

# Crea un mapa centrado en las coordenadas obtenidas
mapa = folium.Map(location=[ubicacion.latitude, ubicacion.longitude], zoom_start=15)

# Agrega un marcador en la ubicaci√≥n
folium.Marker([ubicacion.latitude, ubicacion.longitude], popup=direccion).add_to(mapa)

# Muestra el mapa
mapa
```

```python
# Instala la biblioteca Folium si no est√° instalada
!pip install folium

# Importa la biblioteca Folium
import folium

# Define las coordenadas de los dos puntos
coordenadas_punto1 = (-1.645455, -78.745035)  # Latitud y longitud del primer punto
coordenadas_punto2 = (-1.649985, -78.730797)     # Latitud y longitud del segundo punto
coordenadas_punto3 = (-1.655713, -78.695678)  # Latitud y longitud del primer punto
coordenadas_punto4 = (-1.668838, -78.648573)     # Latitud y longitud del segundo punto
coordenadas_punto5 = (-1.670308, -78.644947)     # Latitud y longitud del segundo punto

# Crea un mapa centrado en la media de las coordenadas de ambos puntos
centro_mapa = [(coordenadas_punto1[0] + coordenadas_punto2[0] + coordenadas_punto3[0] + coordenadas_punto4[0] + coordenadas_punto5[0]) / 5,
               (coordenadas_punto1[1] + coordenadas_punto2[1]  + coordenadas_punto3[1] + coordenadas_punto4[1]  + coordenadas_punto5[1] )/ 5]

mapa = folium.Map(location=centro_mapa, zoom_start=13)

# Agrega un marcador para el primer punto
folium.Marker(coordenadas_punto1, popup='Salida del terminal').add_to(mapa)
folium.Circle(location=coordenadas_punto1, radius=100, color='blue', fill=True, fill_color='blue').add_to(mapa)

# Agrega un marcador para el segundo punto
folium.Marker(coordenadas_punto2, popup='30 pasajeros').add_to(mapa)
folium.Circle(location=coordenadas_punto2, radius=100, color='red', fill=True, fill_color='green').add_to(mapa)

# Agrega un marcador para el tercer punto
folium.Marker(coordenadas_punto3, popup='10 pasajeros').add_to(mapa)
folium.Circle(location=coordenadas_punto3, radius=100, color='red', fill=True, fill_color='green').add_to(mapa)

# Agrega un marcador para el cuarto punto
folium.Marker(coordenadas_punto4, popup='40 pasajeros').add_to(mapa)
folium.Circle(location=coordenadas_punto4, radius=100, color='blue', fill=True, fill_color='blue').add_to(mapa)

# Agrega un marcador para el quinto punto
folium.Marker(coordenadas_punto5, popup='llegada').add_to(mapa)
folium.Circle(location=coordenadas_punto5, radius=100, color='red', fill=True, fill_color='green').add_to(mapa)

# Agrega una l√≠nea entre los dos puntos
nombre_linea = "L√≠nea de buses 14"
folium.PolyLine([coordenadas_punto1, coordenadas_punto2, coordenadas_punto3, coordenadas_punto4, coordenadas_punto5], color='green', weight=2.5, opacity=2.5, tooltip=nombre_linea).add_to(mapa)

# Muestra el mapa
mapa
```
